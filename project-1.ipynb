Visualization of COVID-19 Data

```python

# Group data by date
daily = df.groupby("Date")["Confirmed"].sum()

# Moving average (smooth curve)
rolling = daily.rolling(window=7).mean()

plt.figure()
plt.plot(daily)
plt.plot(rolling)

plt.title("COVID Cases Trend with Moving Average")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.show()
```

```python
top_countries = df.groupby("Country")["Confirmed"].sum().sort_values(ascending=False).head(10)

plt.figure()
top_countries.plot(kind='bar')

plt.title("Top 10 Countries by Cases")
plt.xlabel("Country")
plt.ylabel("Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
df["Confirmed"].plot(kind='hist',bins=40,rwidth=0.8)

plt.title("Distribution of Confirmed Cases")
plt.xlabel("Cases")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1
df.groupby("Date")["Confirmed"].sum().plot(ax=axes[0,0])
axes[0,0].set_title("Confirmed Cases")

# Plot 2
df.groupby("Date")["Deaths"].sum().plot(ax=axes[0,1])
axes[0,1].set_title("Deaths")

# Plot 3
df.groupby("Date")["Recovered"].sum().plot(ax=axes[1,0])
axes[1,0].set_title("Recovered")

# Plot 4
df.groupby("Date")["Active"].sum().plot(ax=axes[1,1])
axes[1,1].set_title("Active Cases")

plt.tight_layout()
plt.show()
```

```python
totals = df[["Confirmed","Recovered","Deaths"]].sum()

plt.figure()
totals.plot(kind='pie', autopct='%1.1f%%')

plt.title("Overall COVID Distribution")
plt.ylabel("")
plt.show()
```

```python


plt.figure()
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)

plt.title("Correlation Between Variables")
plt.show()
```
